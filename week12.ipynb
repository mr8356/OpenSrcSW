```python
#서울의 일교차
import csv
f= open('Seoul.csv' , 'r' , encoding = 'cp949')
data = csv.reader(f)
next(data)
#일교차
max_diff = [0 , '']
#최저기온
low_temp = [99 , '']
#최고기온
high_temp = [0, '']
for row in data:
    h = float(row[4])
    l = float(row[3])
    diff = h-l
    if h>high_temp[0]:
        high_temp = [h , row[0]]
    if l<low_temp[0]:
        low_temp = [l , row[0]]
    if diff>max_diff[0]:
        max_diff = [diff , row[0]]
print('일교차가 가장 큰 날' , max_diff[1])
print('최저기온이 가장 낮은 날' , low_temp[1])
print('최고기온이 가장 높은 날' , high_temp[1])
```
