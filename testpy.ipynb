```python
import matplotlib.pyplot as plt

xdata = [1,2,3,4]
ydata = [10, 5, 37, 50]

plt.plot(xdata , ydata)
plt.show()

plt.bar(xdata , ydata)
plt.show()

plt.barh(xdata , ydata)
plt.show()
```

```python
import csv
f = open('Seoul.csv'  ,encoding = "cp949")
data = csv.reader(f)
next(data)
result = []
for row in data:
    if row[-1]!='':
        if int(row[0].split('.')[1])==3 and int(row[0].split('.')[2])==5:
            result.append(float(row[-1]))
print(result)
```
